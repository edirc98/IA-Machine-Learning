```python
import pandas as pd

df = pd.read_csv("..\datos\Advertising.csv")
df.head()
```

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np


#REGRESION LINEAL TV VS VENTAS
datosX = df['TV']
datosY = df['sales']
slope, b, r_value, p_value, std_err = stats.linregress(datosX, datosY)
#Line Equation
print("y = "+ str(slope) + " * x + " + str(b))
#Squared error
print("Squared Error: " + str(r_value**2))

#Prediction Line
x = np.linspace(0,300,100)
y = slope * x + b

plt.scatter(datosX, datosY)
plt.plot(x, y, c='r')
plt.show()
```

```python
#MULTIVARIABLE REGRESION
import statsmodels.api as sm

x = df[['TV', 'radio', 'newspaper']]
y = df['sales']

#Agregamos la intersección en Y
X = sm.add_constant(x)
est = sm.OLS(y, x).fit()
est.summary()
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

X = df[['TV', 'radio', 'newspaper']]

#.values convierte de DataFrame a Numpy array
#Si queremos obtener el precio estimado, hay que cometar esta línea
X[['TV', 'radio', 'newspaper']] = scale.fit_transform(X[['TV', 'radio', 'newspaper']].values)
est = sm.OLS(y, X).fit()
est.summary()
```
